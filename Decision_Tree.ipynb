```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("cancer_dataset.csv")
df
```

```python
X=df.iloc[:,2:]
y=df.iloc[:,1]
y.value_counts()
```

```python
X=X.dropna(axis=1)
X
```

Cross_val_score

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

```python
model=DecisionTreeClassifier()
score=cross_val_score(model,X,y,cv=10)
print(score)
```

Hold One Out

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=100)
dt=DecisionTreeClassifier()
dtmodel=dt.fit(X_train,y_train)
```

```python
dt.score(X_train,y_train)
```

```python
hoo_result=dtmodel.score(X_test,y_test)
print("The accuracy score id for hold one out method:",hoo_result)
```

K-Fold

```python
from sklearn.model_selection import KFold
kf=KFold(n_splits=5)
kfold_score=cross_val_score(dt,X,y,cv=kf)
print("The cross validation score of k fold is",kfold_score)
kfold_score_mean=kfold_score.mean()
```

```python
print("The minimum accuracy from k fold cv is",min(kfold_score))
print("The maximum accuracy from k fold cv is",max(kfold_score))
print("The mean cross validation k fold method with 5 folds is",kfold_score_mean)
```

```python
Kf2=KFold(n_splits=5)
acc_score_kfold=[]
for train_ind,test_ind in Kf2.split(X,y):
    X_train,X_test=X.iloc[train_ind,:],X.iloc[test_ind,:]
    y_train,y_test=y[train_ind],y[test_ind]
```

```python
dt.fit(X_train,y_train)
pred_values=dt.predict(X_test)
```

```python
acc=accuracy_score(pred_values,y_test)
acc_score_kfold.append(acc)
```

```python
avg_acc_score=sum(acc_score_kfold)/5
print('accuracy of each fold:{}'.format(acc_score_kfold))
print('avg accuracy :{}'.format(avg_acc_score))
```
