```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
social=pd.read_csv("Social_Network_Ads.csv")
social
```

```python
social.isna().sum()
```

```python
X=social.iloc[:,[2,3]].values
y=social.iloc[:,4].values
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
#Featuring scaling of independent variables for accurate prediction
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.tree import DecisionTreeClassifier
cf=DecisionTreeClassifier(random_state=6)
cf.fit(X_train,y_train)
```

```python
y_pred=cf.predict(X_test)
y_pred
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
print("Acurracy of the model:%".format(accuracy_score(y_pred,y_test)*100))
```

```python
#accuracy
from sklearn import metrics
print(metrics.accuracy_score(y_test,y_pred))
```

```python
accuracy=[]
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

for i in range(1,10):
    model=DecisionTreeClassifier(max_depth=i,random_state=0)
    model.fit(X_train,y_train)
    pred=model.predict(X_test)
    score=accuracy_score(y_test,pred)
    accuracy.append(score)
```

```python
plt.figure(figsize=(12,6))
plt.plot(range(1,10), accuracy, color='red', linestyle='dashed', marker='o', markerfacecolor='blue', markersize=10)
plt.title('Finding max_depth')
plt.xlabel('Prediction')
plt.ylabel('Score')
```

```python
from sklearn.tree import DecisionTreeClassifier
cf1=DecisionTreeClassifier(max_depth=2, random_state=6)
cf1.fit(X_train, y_train)
pred=cf1.predict(X_test)
```

```python
#accuracy
from sklearn import metrics
print(metrics.accuracy_score(y_test,pred))
```
